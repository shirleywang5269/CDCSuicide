```python
import pandas as pd
df_full_data_1 = pd.read_csv(r'suicide2009-2013.txt', sep='\t')
df_full_data_1.head()
```

```python
df_full_data_1['Age Adjusted Rate'] = pd.to_numeric(df_full_data_1['Age Adjusted Rate'], errors='coerce')
# no_quotes['Age Adjusted Rate'].max()
# no_quotes['Age Adjusted Rate'].min()
df_full_data_1 = df_full_data_1.sort_values('Age Adjusted Rate',ascending=False)

#This is the way to show deaths occured per year for every 100000 persons
df_full_data_1['Suicide_Death_Rate']=(df_full_data_1['Deaths']/df_full_data_1.Population)*100000
df_full_data_1[:5]
```

$$ Crude Rate Upper Limit =  (100000/n)*(d+ 1.96 \sqrt{d}) ; d=death; n=population$$

$$ Age Adjusted Rate Upper Limit =  r+(1.96*r/ \sqrt{n}) ; r=age rate; n=death$$

### https://www.health.pa.gov/topics/HealthStatistics/Statistical-Resources/UnderstandingHealthStats/Documents/Confidence_Intervals_for_a_Crude_Rate.pdf
### https://www.health.pa.gov/topics/HealthStatistics/Statistical-Resources/UnderstandingHealthStats/Documents/Confidence_Intervals_for_an_Age_Adjusted_Rate.pdf

## Generating 16 intervals of Age adjusted Rate

```python
import numpy as np
def New_Adjusted_Rate(n,s):
    return (25*np.sqrt(n)*s/(25*np.sqrt(n)+49))
df_full_data_1['Calculated_Age_Adjusted_Rate']=df_full_data_1.apply(lambda x: New_Adjusted_Rate(x.Deaths,x['Age Adjusted Rate Upper 95% Confidence Interval']),axis=1)
# df_full_data_1['bin']=pd.cut(df_full_data_1['Calculated_Age_Adjusted_Rate'], bins=16)
# df_full_data_1
df_full_data_1['binned']=pd.cut(df_full_data_1['Calculated_Age_Adjusted_Rate'], bins=10).astype(str)

df_full_data_1
```

## Association with FIPS 

```python
import geopandas as gpd
dfg = gpd.read_file('us-counties.json')
# print(type(dfg))

dfg['id'] = dfg['id'].apply(lambda x: x.zfill(5))
df_full_data_1['County Code']=df_full_data_1['County Code'].astype(str)
df_full_data_1['County Code']= df_full_data_1['County Code'].apply(lambda x: x.zfill(5))

df_2013 = df_full_data_1[(df_full_data_1.Year == 2013)]
df_2013.head()

years = sorted(df_full_data_1['Year'].unique().tolist())
years
bins = df_2013['binned'].unique()
bins
```

```python
df_2013
```

```python
for yr in years:
    df_single_year = df_full_data_1[(df_full_data_1.Year == yr)]
    merged = dfg.merge(df_single_year, right_on='County Code', left_on='id' )
    for bin in bins:
        print(yr,bin)
        geo_layer = merged[(merged['binned'] == bin)] # geo
        try:
            geo_layer.to_file("{0}/{1}.geojson".format(yr, bin), driver='GeoJSON')
        except:
            print('err')
merged = dfg.merge(df_2013, right_on='County Code', left_on='id' )
merged = merged.drop(columns=['Notes','Year Code'])
merged.head()
merged.to_csv('age_adjusted_death_rate_suicide.csv')
```
